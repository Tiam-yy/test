```python
import os
```

### 获取当前工作路径

```python
os.getcwd()
```

传入任意一个path，返回的是该路径下所有文件和目录组成的列表

```python
pa = r'E:\vscode projects'
os.listdir(pa)
```

传入任意一个path，深层次遍历指定路径下所有子文件夹，返回的函数一个由路径、文件夹列表、文件列表组成的元组

```python
pa = r'E:\vscode projects'
for path ,dirs, files in os.walk(pa):
    print(path)
    print(dirs)
    print(files)
    print()
```

传入一个path，判断指定路径下的目录是否存在。

```python
pa1 = r'E:\vscode projects'
os.path.exists(pa1)
```

传入一个path路径，创建单层（单个）文件夹；如果文件夹已经存在，就会报错。

```python
pa2 = os.getcwd() + r'\test'
os.mkdir(pa2)
```

传入一个path，生成一个递归的文件夹

```python
pa3 = os.getcwd() + r'\test1\test2'
os.makedirs(pa3)
```

传入一个path，删除指定目录下的文件夹
attention：只能删除空文件夹，删除非空文件夹会报错

```python
pa4 = os.getcwd() + r'\test'
os.rmdir(pa4)
# 从给定路径开始，自底向上递归删除所有变为空的父目录
os.removedirs(os.getcwd() + r'\test1\test2')
```

传入两个path，将该路径拼接起来，形成一个新的完整路径

```python
# os.path.join()
pa = os.getcwd()
lis = ['a.jpg', 'b.jpg', 'c.jpg']
for i in lis:
    x = os.path.join(pa,i)
    print(x)
```

传入一个完整的path，将其拆分为父目录路径和文件名两个部分；将path分割成目录和文件名并以元组方式返回

```python
pa = os.getcwd() + r'\a.jpg'
# os.mkdir(pa)
os.path.split(pa)
```

也可以只返回父目录路径或者只返回文件名

```python
pa = os.getcwd() + r'\a.jpg'
print(os.path.dirname(pa))
print(os.path.basename(pa))
```

传入一个完整的文件路径，判断它是否是文件夹/文件

```python
file_list = os.listdir()
for i in file_list:
    if os.path.isdir(i):
        print('dir',i)
    elif os.path.isfile(i):
        print('file',i)
```

返回当前操作系统的路径分割符

```python
os.path.sep
```

返回path的绝对路径

```python
os.path.abspath('.')
```

传入一个完整的文件路径，返回该文件的大小（不能是文件夹） 单位：字节byte

```python
pa = os.getcwd() + '\\2.1.ipynb'
os.path.getsize(pa)
```
